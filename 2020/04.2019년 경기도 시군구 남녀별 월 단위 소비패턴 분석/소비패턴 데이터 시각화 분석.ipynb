```python
df_sex = df.groupby(['가맹점업종명','성별코드']).sum().unstack('성별코드')
```

```python
df_sex.columns
```

```python
df_sex['total'] = df_sex['총결제금액', 'F'] + df_sex['총결제금액', 'M']
```

```python
df_sex_sorted = df_sex.sort_values(by=['total'], ascending=False)
df_sex_sorted = df_sex_sorted.drop(['total'], axis = 1)
```

```python
df_sex_sorted.plot(kind='bar',figsize=(12,10), title = "경기도 남녀별 가맹점업종별 소비 패턴")
```

```python
df_bu = df[df['시군구명'] == '부천시']
```

```python
df_bu_new = df_bu.groupby(['가맹점업종명','성별코드']).sum().unstack('성별코드')
```

```python
df_bu_new['total'] = df_bu_new['총결제금액', 'F'] + df_bu_new['총결제금액', 'M']
df_bu_new = df_bu_new.sort_values(by=['total'], ascending=False)
df_bu_new = df_bu_new.drop(['total'], axis = 1)
```

```python
df_bu_new[0:10].plot(kind='bar',figsize=(12,10), title = "부천시 남녀별 가맹점업종별 소비 패턴")
```

```python
df_hwa = df[df['시군구명'] == '화성시']
df_hwa_new = df_hwa.groupby(['가맹점업종명','성별코드']).sum().unstack('성별코드')
df_hwa_new['total'] = df_hwa_new['총결제금액', 'F'] + df_hwa_new['총결제금액', 'M']
df_hwa_new = df_hwa_new.sort_values(by=['total'], ascending=False)
df_hwa_new = df_hwa_new.drop(['total'], axis = 1)
```

```python
df_hwa_new[0:10].plot(kind='bar',figsize=(12,10), title = "화성시 남녀별 가맹점업종별 소비 패턴")
```
