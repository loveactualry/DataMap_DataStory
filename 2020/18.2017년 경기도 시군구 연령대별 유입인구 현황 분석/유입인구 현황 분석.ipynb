```python
import pandas as pd
import seaborn as sns
import os
```

```python
# matplotlib 설정하기
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.style.use("ggplot")
```

```python
data_folder = "./유입인구 현황 데이터/"
```

```python
all_df = []

data_path_list = os.listdir(data_folder)

for i in range(0, len(data_path_list)):
    
    if i < 12:
    
        file_name = data_path_list[i]
        print(file_name)

        df = pd.read_csv(data_folder+file_name, encoding = 'utf-8')

        all_df.append(df)

merged_df = pd.concat(all_df, ignore_index=True, sort=True)
```
