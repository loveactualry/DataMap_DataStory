```python
question_dic = {}

for question in list(정규과정설문_df['질문명']):
    
    if question in question_dic:
        
        question_dic[question] += 1
    else:
        question_dic[question] = 1
```

```python
question_dic
```

```python
sorted(question_dic, key=lambda key_value: key_value, reverse=True)
```

```python
import operator
dict( sorted(question_dic.items(), key=operator.itemgetter(1),reverse=True))
```

```python
dict( sorted(question_dic.items(), key=operator.itemgetter(1),reverse=True))
```

```python
top_question_list = sorted(question_dic, key=question_dic.get, reverse=True)[0:11]
```

```python
top_question_list
```

```python
정규과정설문_df[정규과정설문_df['질문명'] == top_question_list[1]]
```

```python
정규과정설문_df[정규과정설문_df['질문명'] == top_question_list[0]]
```

```python
for question in top_question_list:
    
    plt.figure(figsize=(12, 8))
    sns.countplot(y='답변항목명', data=정규과정설문_df[정규과정설문_df['질문명'] == question])
    plt.title(question+" 에 대한 답변 통계")
```
